```python
import glob
import pandas as pd
import numpy as np
for table in glob.glob("*.csv"):
    print(f'{"*"*32} {table} Summary: {"*"*32}')
    data = pd.read_csv(table)
    print (data.describe())
    print(f'{"*"*8} Data Types: {"*"*8}')   
    print(data.dtypes)
    print(f'{"*"*8} Number of Unique Values: {"*"*8}')     
    print(data.nunique())
    print(f'{"*"*8} Null Values: {"*"*8}')
    print(data.isna().sum())
    print("*"*100)
```
