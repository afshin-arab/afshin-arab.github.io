```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
import numpy as np
```

```python
data = pd.read_csv("data.csv")
```

```python
data.head()
```

```python
data.shape
```

```python
X = data.drop("worldwide_gross_usd", axis=1)
y = data["worldwide_gross_usd"] #worldwide_gross_usd
```

```python
plt.figure(figsize=(12,8))
plt.scatter(X,y,alpha = .5)
plt.title("Movie Revenue vs. Movie Bidget Scatterplot")
plt.xlabel("Production Budget")
plt.ylabel("Worldwide Gross")
plt.ylim(0,3e9)
plt.xlim(0,4.5e8)
```

```python
regression = LinearRegression()
```

```python
regression.fit(X, y)
```

```python
regression.coef_
```

```python
regression.intercept_
```

```python
regression.score(X,y)
```

```python
plt.figure(figsize=(12,8))
plt.scatter(X,y,alpha = .5)
plt.plot(X,regression.predict(X),color = 'r', linewidth = 4)
plt.title("Movie Revenue vs. Movie Bidget Scatterplot")
plt.xlabel("Production Budget")
plt.ylabel("Worldwide Gross")
plt.ylim(0,3e9)
plt.xlim(0,4.5e8)
```
